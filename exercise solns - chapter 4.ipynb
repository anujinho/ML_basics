# Ex.4.2 

```python
samples=100
x1, y1 = np.random.multivariate_normal([5,5], [[2,0], [0,2]], samples).T
x2, y2 = np.random.multivariate_normal([10,10], [[2,0], [0,2]], samples).T
a1 = np.vstack((x1,y1)).T; a2 = np.vstack((x2,y2)).T
a = np.concatenate((a1,a2))
a[:,1] = a[:,1]
labels = np.concatenate((np.zeros(samples), np.ones(samples))).T
a = pr.prdataset(a, labels)
ws = pr.svc(a)
#ws.train(a)
pr.scatterd(a)
pr.plotc(ws, colors='red')
```

```python
samples=100
x1, y1 = np.random.multivariate_normal([5,5], [[2,0], [0,2]], samples).T
x2, y2 = np.random.multivariate_normal([10,10], [[2,0], [0,2]], samples).T
a1 = np.vstack((x1,y1)).T; a2 = np.vstack((x2,y2)).T
a = np.concatenate((a1,a2))
a[:,1] = 100*a[:,1] # stretching y axis
labels = np.concatenate((np.zeros(samples), np.ones(samples))).T
a = pr.prdataset(a, labels)
ws = pr.svc(a)
#ws.train(a)
pr.scatterd(a)
pr.plotc(ws, colors='red')
```

Stretching the Y-axis results in an extremely overfitted and unstable boundary. This could be because the decision boundary adjusts its parameters based on each training point without exactly taking into consideration its distribution (class distribution). Lambda vals are non-zero for support vectors and the params are essentially decided by only the support vectors and thus if the features aren't scaled well enough, the boundary follows the poorly scaled vector directions of the individual support vectors (linear transform of them). This reasoning can be extended for non-linear kernels too.

```python
samples=100
x1, y1 = np.random.multivariate_normal([5,5], [[2,0], [0,2]], samples).T
x2, y2 = np.random.multivariate_normal([10,10], [[2,0], [0,2]], samples).T
a1 = np.vstack((x1,y1)).T; a2 = np.vstack((x2,y2)).T
a = np.concatenate((a1,a2))
a[:,1] = 100*a[:,1] # stretching y axis
labels = np.concatenate((np.zeros(samples), np.ones(samples))).T
a = pr.prdataset(a, labels)
ws = pr.svc(a); wl = pr.ldc(a)
#ws.train(a)
pr.scatterd(a)
pr.plotc(ws, colors='red')
pr.plotc(wl, colors='yellow')
```

LDA classifier is forced to construct the covariance matrix of the classes and thus takes into account the distribution of the classes/data. Thus its immune to feature scaling. 

# Ex.4.7

```python
a = pr.gendatb([500,500])
w1 = pr.svc(a, ('rbf', 0.1, 1))
w2 = pr.svc(a, ('rbf', 10, 1))
pr.scatterd(a)
pr.plotc(w1, colors='red')
pr.plotc(w2, colors='yellow')
```

a. Small width causes the classifier to over fit due to its nature of capturing every small intricacy within the class's data's support vector's distribution.

```python
w3 = pr.parzenc(a, 0.1)
w4 = pr.parzenc(a, 10)
pr.scatterd(a)
pr.plotc(w3, colors='green')
pr.plotc(w4, colors='red')
```

b. Parzen Density estimator still uses bayesian decision theory to draw the classifier. Thus for a greater width, it's a straight line due to underfiting of the density estimate of each class.

c. Kernel SVC uses only the support vectors to calculate the params of decision boundary whereas the KDE uses every datum to estimate density first and then use Bayesian decision theory to arrive at a decision boundary.

# Ex.4.9 

```python
help(pr.proxm)
```

```python
samples = 500
x = np.random.normal(0.5, 0.5, size=(samples,2))
y = np.array(50*np.sin(x[:,0])*np.sin(x[:,1]), ndmin=2).T
e = np.random.normal(0.5, 0.5, size=(samples,1))
y = y + e
a = pr.gendatr(x,y)
wkr = pr.proxm(('rbf', 4))*pr.ridger(0.01)
wkr = a*wkr
pr.scatterr(a)
pr.plotr(wkr, color='red')
b = wkr.eval(a)
e = pr.testr(b)
print('Error: ', e)
```


# Ex.4.10

```python
samples = 40; split = int(0.8*samples)
x = np.random.uniform(low=0, high=1, size=(samples,1))
e = np.random.normal(0.5, 0.5, size=(samples,1))
y = x**2 + e
atrain = pr.gendatr(x[0:split], y[:split])
atest = pr.gendatr(x[split:], y[split:])
w = pr.linearr(atrain, 5)
pr.scatterr(atrain)
pr.plotr(w, color='red')
b = w.eval(atest)
e = pr.testr(b)
print('Error: ', e)
```

The variance increases as the polynomial's degree increases and bias decreases.
